```python
# Death Over Data
death_over = delivery[delivery['over']>15]
death_over
```

```python
# No. of balls batsman play in death over
death_over.groupby('batsman')['batsman_runs'].count()
```

```python
# No. of balls batsman play in death over is more than 200
d_over= death_over.groupby('batsman')['batsman_runs'].count()
x = d_over>200
d_over[x]
```

```python
batsman_dover = d_over[x].index.tolist()
```

```python
delivery[delivery['batsman'].isin(batsman_dover)]
```
