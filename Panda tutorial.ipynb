```python
def run_scored(batsman_name):
    batsman_data = delivery[delivery['batsman']==batsman_name]  ## provide data in true or false and to convert it in database put in delivery
    result = batsman_data.groupby('bowling_team')['batsman_runs'].sum().sort_values().tail(4)
    return result
```

```python
run_scored('Yuvraj Singh')
```
